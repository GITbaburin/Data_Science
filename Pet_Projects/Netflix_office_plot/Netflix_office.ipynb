```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/office_episodes.csv')
df.head()
```

```python
df[df['viewership_mil'] > 20]
```

```python
def func(x):
    if x < 0.25:
        return 'red'
    elif 0.25 <= x < 0.50:
        return 'orange'
    elif  0.50 <= x < 0.75:
        return 'lightgreen'
    elif  0.75 <= x:
        return 'darkgreen'

df['rating_cat'] = df['scaled_ratings'].apply(func)
```

```python
def func_size(x):
    if x == True:
        return 250
    else:
        return 25

df['dot_size'] = df['has_guests'].apply(func_size)
```

```python
df['dot_size'].unique()
```

```python
fig = plt.figure(figsize=(15, 10))
x = df['episode_number']
y = df['viewership_mil']
z = df['rating_cat']
s = df['dot_size']
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.scatter(x, y, c=z, s=s)
plt.show()
```
